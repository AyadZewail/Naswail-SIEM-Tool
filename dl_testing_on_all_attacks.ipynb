```python
import time
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from tqdm import tqdm
import dask.dataframe as dd
from sklearn.preprocessing import MinMaxScaler
import copy
from scipy import stats
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, auc
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import joblib
import pandas as pd
import joblib
import random
# Dataset & Model
class NumericDataset(Dataset):
    def __init__(self, data):
        self.data = torch.tensor(data, dtype=torch.float32)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx], self.data[idx]
# Load and Preprocess Data

random.seed(42)  # set seed once outside
df_big = pd.read_csv(
    "/content/best-cic-2018-merged-bening-and-attack-seperated/better_benign_no_label.csv",
    skiprows=lambda i: i > 0 and random.random() > 0.15  # keep ~15% rows
)
scaler = MinMaxScaler()   #Scale the data to avoid sensitvity to outliers
df_big = scaler.fit_transform(df_big)

# Save the fitted scaler for future use
joblib.dump(scaler, "#base_MLP_scaler.pkl")
#/////////////////////////////////////////////////////////////////////////////
#n_samples = int(1 * len(df_big))
#rng = np.random.default_rng(seed=42)   # reproducible like pandas' random_state
#idx = rng.choice(len(df_big), size=n_samples, replace=False)
df = df_big#df_big[idx] #Use only 10 percent

X_train, X_test = train_test_split(df, test_size=0.2, random_state=42)
#  Training Setup
train_dataset = NumericDataset(X_train)
test_dataset = NumericDataset(X_test)
g = torch.Generator()
g.manual_seed(42)
train_loader = DataLoader(train_dataset, batch_size=1880, shuffle=True,generator=g)
test_loader = DataLoader(test_dataset, batch_size=1280, shuffle=False)




class ImprovedAutoencoder(nn.Module):
    def __init__(self, input_dim):
        super(ImprovedAutoencoder, self).__init__()
        bottleneck_dim = 4#max(8, input_dim // 6)

        self.encoder = nn.Sequential(
            nn.Linear(76, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Dropout(0.1),

            nn.Linear(512, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(0.1),

            nn.Linear(128, 32),
            nn.BatchNorm1d(32),
            nn.ReLU(),
            nn.Dropout(0.1),

            nn.Linear(32, 32),
            nn.BatchNorm1d(32),
            nn.ReLU(),
            nn.Dropout(0.1),

            nn.Linear(32, 32),
            nn.BatchNorm1d(32),
            nn.ReLU(),
            nn.Dropout(0.1),

            nn.Linear(32, 16),
            nn.BatchNorm1d(16),
            nn.ReLU(),
            nn.Dropout(0.1),

            nn.Linear(16, 8)
        )

        self.decoder = nn.Sequential(
            nn.Linear(8, 16),
            nn.BatchNorm1d(16),
            nn.ReLU(),
            nn.Dropout(0.1),

            nn.Linear(16, 32),
            nn.BatchNorm1d(32),
            nn.ReLU(),
            nn.Dropout(0.1),

            nn.Linear(32, 32),
            nn.BatchNorm1d(32),
            nn.ReLU(),
            nn.Dropout(0.1),

            nn.Linear(32, 32),
            nn.BatchNorm1d(32),
            nn.ReLU(),
            nn.Dropout(0.1),

            nn.Linear(32, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(0.1),

            nn.Linear(128, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Dropout(0.1),

            nn.Linear(512, 76)

        )


    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded
# Huber loss is more stable
def robust_loss(output, target, delta=1.0):
    residual = torch.abs(output - target)
    return torch.where(
        residual < delta,
        0.5 * residual**2,
        delta * residual - 0.5 * delta**2
    ).mean()
#  Training Setup
train_dataset = NumericDataset(X_train)
test_dataset = NumericDataset(X_test)
g = torch.Generator()
g.manual_seed(42)
train_loader = DataLoader(train_dataset, batch_size=1882, shuffle=True,generator=g)
test_loader = DataLoader(test_dataset, batch_size=1280, shuffle=False)


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.manual_seed(42)

model = ImprovedAutoencoder(input_dim=76).to(device)
total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"Total trainable parameters: {total_params:,}")
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

NUM_EPOCHS = 10
train_losses = []
eval_losses = []
best_eval=99999999
patience=3
bad=0
best_model=None
#  Training Loop
for epoch in range(NUM_EPOCHS):
    #time.sleep(0.02)
    model.train()
    total_train_loss = 0
    loop = tqdm(train_loader, desc=f"Epoch {epoch+1}/{NUM_EPOCHS}", leave=False)
    for batch_x, _ in loop:

        batch_x = batch_x.to(device)
        output = model(batch_x)
        loss = robust_loss(output, batch_x)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        total_train_loss += loss.item()
        loop.set_postfix(loss=loss.item())

    avg_train_loss = total_train_loss / len(train_loader)
    train_losses.append(avg_train_loss)

    # Evaluation
    model.eval()
    total_eval_loss = 0
    with torch.no_grad():
        for batch_x, _ in test_loader:
            batch_x = batch_x.to(device)
            output = model(batch_x)
            loss = criterion(output, batch_x)
            total_eval_loss += loss.item()
    avg_eval_loss = total_eval_loss / len(test_loader)
    eval_losses.append(avg_eval_loss)
    if avg_eval_loss<best_eval :
        print("yay")
        best_eval=avg_eval_loss
        #best_model = copy.deepcopy(model.state_dict())
        bad=0
    else:
        bad+=1
    if bad>=patience:
        break


    print(f"Epoch {epoch+1}/{NUM_EPOCHS}, Train Loss: {avg_train_loss:.4f}, Eval Loss: {avg_eval_loss:.4f}")




#  Final Evaluation


model.eval()
reconstruction_errors = []
num=0
with torch.no_grad():
    for batch_x, _ in tqdm(test_loader, desc="Evaluating"):
        num+= 1
        batch_x = batch_x.to(device)
        output = model(batch_x)
        loss = criterion(output,batch_x) # MSE
        reconstruction_errors.append(loss.cpu().numpy())
reconstruction_errors = np.array(reconstruction_errors)
print("number of batches:",num)

# Threshold from benign (test) distribution
threshold = np.percentile(reconstruction_errors, 94)


mean_error = reconstruction_errors.mean()

# Classify anomalies (all benign here that are anomalies are actually false positives)
anomalies = reconstruction_errors > threshold
false_positives = anomalies.sum()

# Print results
print(f"Mean reconstruction error: {mean_error:.6f}")
print(f"Anomaly threshold (99th percentile): {threshold:.6f}")
print(f"False positives: {round(false_positives/len(anomalies)* 100)}% out of {len(anomalies)} samples")




print(f"Detected {anomalies.sum()} anomalies out of {len(anomalies)} samples")

#  Save Model
torch.save(best_model, 'mlp_best_numeric_autoencoder.pth')


epochs = range(1, len(train_losses) + 1)

epochs = range(1, len(train_losses) + 1)

plt.figure(figsize=(10, 5))
plt.plot(epochs, train_losses, label='Train Loss')
plt.plot(epochs, eval_losses, label='Eval Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training vs Evaluation Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

#  Plot 2: Reconstruction Errors

plt.figure(figsize=(8, 5))
plt.hist(reconstruction_errors, bins=100, color='skyblue', alpha=0.8)
plt.axvline(threshold, color='red', linestyle='--', label=f"Threshold = {threshold:.3f}")
plt.xlabel("Reconstruction Error")
plt.ylabel("Frequency")
plt.title("Distribution of Reconstruction Errors")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
# Mean and Std reconstruction error
mean_error = reconstruction_errors.mean()
std_error = reconstruction_errors.std()
print(f"Mean reconstruction error: {mean_error:.6f}")
print(f"Std reconstruction error:  {std_error:.6f}")
print(f"Anomaly threshold (95th percentile): {threshold:.6f}")

print(f"False positives: {round(false_positives/len(anomalies)*100)}% out of {len(anomalies)} samples")

#///////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////
#///////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////
print("testing for fp/////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////")
#testing for fp:
#testing for fp:
fp_rate=0
#df = pd.read_csv('/content/benign-no-label/benign_no_label.csv')




#df.to_csv('attack_no_label2.csv', index=False)


#  Load and Preprocess Malicious & Benign Data

full_benign = df_big


#device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.eval()
full_benign_loader = DataLoader(NumericDataset(full_benign), batch_size=128, shuffle=False)

#  Get Reconstruction Errors

def get_reconstruction_errors(loader):
        errors = []
        with torch.no_grad():
            for batch_x, _ in(loader):
                batch_x = batch_x.to(device)
                output = model(batch_x)

                loss = torch.mean((output - batch_x)**2, dim=1)
                errors.extend(loss.cpu().numpy())
        return np.array(errors)


benign_errors = reconstruction_errors
malicious_errors = get_reconstruction_errors(full_benign_loader)#malicous as in its benign but unseen

#  Threshold and Classification
threshold = np.percentile(benign_errors, 94)

benign_preds = benign_errors >threshold
malicious_preds=malicious_errors >threshold

# Metrics
tp = np.sum(malicious_preds)
fp = np.sum(benign_preds)
fn = len(malicious_errors) - tp
tn = len(benign_errors) - fp

precision = tp / (tp + fp + 1e-8)
recall = tp / (tp + fn + 1e-8)
f1 = 2 * precision * recall / (precision + recall + 1e-8)
accuracy = (tp + tn) / (len(benign_errors) + len(malicious_errors))



fp_rate=recall
print(f"false positve rate:{fp_rate}")

#//////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////
#////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////
print("actual testing/////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////")

#testing on all attacks
# Create a doc at the start
from io import StringIO
from docx import Document
from docx.shared import Inches
import os
report_doc = Document()
report_doc.add_heading("Anomaly Detection Report", 0)

# Directory for saving temporary plots
os.makedirs("plots", exist_ok=True)
attacks=os.listdir("/content/best-cic-2018-merged-bening-and-attack-seperated/attacks")
#attacks=["SSH-Bruteforce.csv","FTP-BruteForce.csv","DoS_attacks-Slowloris.csv","DoS attacks-GoldenEye.csv","DDOS_attack-LOIC-UDP.csv","DDOS_attack-HOIC.csv","Brute_Force_-XSS.csv","Brute_Force_-Web.csv","Bot.csv","sql_injection.csv"]
for att in attacks:

    df = pd.read_csv('/content/best-cic-2018-merged-bening-and-attack-seperated/attacks/'+att)
    #df = df.sample(frac=0.65, random_state=42).reset_index(drop=True)
    print(len(df))
    # Drop rows where ' Label' is  'BENIGN'
    df_attack = df[df['Label'] != 'Benign'].copy()

    print(f"Remaining rows: {len(df_attack)}")

    # Optional: save to a new CSV
    df_attack.to_csv('attack_only2.csv', index=False)
    print("saved")
    df = df.drop(columns=['Label'])
    # Replace inf and -inf with NaN
    import numpy as np
    df = df.replace([np.inf, -np.inf], np.nan)

    # Drop rows with any NaN
    df = df.dropna()

    df.to_csv('attack_no_label2.csv', index=False)
    #########################################

    #  Load and Preprocess Malicious & Benign Data
    malicious_df = pd.read_csv("attack_no_label2.csv")

    malicious_df = malicious_df.select_dtypes(include='number').dropna()
    #malicious_df = malicious_df.sample(frac=0.10, random_state=42).reset_index(drop=True)


    # Load saved scaler from training
    scaler = joblib.load("#base_MLP_scaler.pkl")

    # Use the training scaler for both benign & malicious test data
    #benign_scaled = scaler.transform(benign_df)
    malicious_scaled = scaler.transform(malicious_df)


    print(len(malicious_scaled))

    # ------------- Setup -------------
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    #input_dim = malicious_scaled.shape[1]

    # Load model
    #model = ImprovedAutoencoder(input_dim=input_dim).to(device)
    #model.load_state_dict(torch.load("models/regular_best_numeric_autoencoder.pth", map_location=device))

    model.eval()

    # Dataloaders
    #benign_loader = DataLoader(NumericDataset(benign_scaled), batch_size=64, shuffle=False)
    subset_data = malicious_scaled[:min(2093, len(malicious_scaled))]
    malicious_loader = DataLoader(NumericDataset(subset_data), batch_size=128, shuffle=False)


    #malicious_loader = DataLoader(NumericDataset(malicious_scaled), batch_size=128, shuffle=False)
    print(len(malicious_loader))
    #  Get Reconstruction Errors

    def get_reconstruction_errors(loader):
        errors = []
        with torch.no_grad():
            for batch_x, _ in tqdm(loader):
                batch_x = batch_x.to(device)
                output = model(batch_x)

                loss = torch.mean((output - batch_x)**2, dim=1)
                errors.extend(loss.cpu().numpy())
        return np.array(errors)

    print("Computing reconstruction errors...")
    benign_errors = reconstruction_errors
    malicious_errors = get_reconstruction_errors(malicious_loader)

    # Threshold and Classification
    threshold = np.percentile(benign_errors, 94)

    benign_preds = benign_errors > threshold
    malicious_preds = malicious_errors > threshold

    # Metrics
    tp = np.sum(malicious_preds)
    fp = np.sum(benign_preds)
    fn = len(malicious_errors) - tp
    tn = len(benign_errors) - fp

    precision = tp / (tp + fp + 1e-8)
    recall = tp / (tp + fn + 1e-8)
    f1 = 2 * precision * recall / (precision + recall + 1e-8)
    accuracy = (tp + tn) / (len(benign_errors) + len(malicious_errors))

    # Statistical test
    stat, p_value = stats.mannwhitneyu(malicious_errors, benign_errors, alternative='greater')

    # ROC Curve
    thresholds = np.linspace(min(np.min(benign_errors), np.min(malicious_errors)),
                            max(np.max(benign_errors), np.max(malicious_errors)), 100)
    tpr_list, fpr_list = [], []

    for t in thresholds:
        tpr = np.sum(malicious_errors > t) / len(malicious_errors)
        fpr = np.sum(benign_errors > t) / len(benign_errors)
        tpr_list.append(tpr)
        fpr_list.append(fpr)

    roc_auc = auc(fpr_list, tpr_list)

    # Classification report + Confusion matrix
    y_true = np.array([0] * len(benign_errors) + [1] * len(malicious_errors))
    y_pred = np.concatenate([benign_preds, malicious_preds]).astype(int)

    print("\n" + "="*50)
    print("SKLEARN CLASSIFICATION REPORT")
    print("="*50)
    print(classification_report(y_true, y_pred, target_names=["Benign", "Malicious"], digits=4))

    print("\nConfusion Matrix:")
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Benign", "Malicious"], yticklabels=["Benign", "Malicious"])
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.title("Confusion Matrix")
    plt.tight_layout()
    plt.show()

    #  Final Logs
    print("\n" + "="*50)
    print("ANOMALY DETECTION RESULTS")
    print("="*50)
    print(f"Threshold: {threshold:.4f}")
    print(f"Benign mean error:    {benign_errors.mean():.4f} | Std: {benign_errors.std():.4f}")
    print(f"Malicious mean error: {malicious_errors.mean():.4f} | Std: {malicious_errors.std():.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall:    {recall:.4f}")
    print(f"F1 Score:  {f1:.4f}")
    print(f"Accuracy:  {accuracy:.4f}")
    print(f"p-value (MWU): {p_value:.2e} | Significant: {'Yes' if p_value < 0.05 else 'No'}")
    print(f"AUC: {roc_auc:.4f}")


    #  Visualizations
    plt.figure(figsize=(15, 8))

    plt.subplot(2, 2, 1)
    plt.hist(benign_errors, bins=50, alpha=0.6, label='Benign', color='blue', density=True)
    plt.hist(malicious_errors, bins=50, alpha=0.6, label='Malicious', color='red', density=True)
    plt.axvline(threshold, color='black', linestyle='--', label='Threshold')
    plt.title("Reconstruction Error Distribution")
    plt.legend()
    plt.grid(True)

    plt.subplot(2, 2, 2)
    plt.boxplot([benign_errors, malicious_errors], labels=["Benign", "Malicious"])
    plt.axhline(threshold, color='red', linestyle='--')
    plt.title("Boxplot of Reconstruction Errors")
    plt.grid(True)

    plt.subplot(2, 2, 3)
    plt.scatter(range(len(benign_errors)), benign_errors, s=10, label="Benign", alpha=0.5)
    plt.scatter(range(len(benign_errors), len(benign_errors)+len(malicious_errors)), malicious_errors, s=10, label="Malicious", alpha=0.5, color='red')
    plt.axhline(threshold, color='black', linestyle='--')
    plt.title("Error Scatter Plot")
    plt.legend()
    plt.grid(True)

    plt.subplot(2, 2, 4)
    plt.plot(fpr_list, tpr_list, label=f"AUC = {roc_auc:.3f}", linewidth=2)
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()
    print(f"Recall (malicious) = {recall:.4f}  # or {recall*100:.2f}%")
    og_buffer = StringIO()
    log_buffer = StringIO()
    # Classification Report
    cls_report = classification_report(
        y_true, y_pred, target_names=["Benign", "Malicious"], digits=4
    )
    print("\n" + "="*50, file=log_buffer)
    print("SKLEARN CLASSIFICATION REPORT", file=log_buffer)
    print("="*50, file=log_buffer)
    print(cls_report, file=log_buffer)

    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=["Benign", "Malicious"],
                yticklabels=["Benign", "Malicious"])
    plt.title("Confusion Matrix")
    cm_path = f"plots/{att}_cm.png"
    plt.savefig(cm_path)
    plt.close()

    # Other visualizations (save instead of show)
    fig, axes = plt.subplots(2, 2, figsize=(15, 8))

    # 1. Histograms
    axes[0,0].hist(benign_errors, bins=50, alpha=0.6, label='Benign', color='blue', density=True)
    axes[0,0].hist(malicious_errors, bins=50, alpha=0.6, label='Malicious', color='red', density=True)
    axes[0,0].axvline(threshold, color='black', linestyle='--')
    axes[0,0].set_title("Reconstruction Error Distribution")
    axes[0,0].legend()

    # 2. Boxplot
    axes[0,1].boxplot([benign_errors, malicious_errors], labels=["Benign", "Malicious"])
    axes[0,1].axhline(threshold, color='red', linestyle='--')
    axes[0,1].set_title("Boxplot of Errors")

    # 3. Scatter
    axes[1,0].scatter(range(len(benign_errors)), benign_errors, s=10, alpha=0.5, label="Benign")
    axes[1,0].scatter(range(len(benign_errors), len(benign_errors)+len(malicious_errors)), malicious_errors, s=10, color='red', alpha=0.5, label="Malicious")
    axes[1,0].axhline(threshold, color='black', linestyle='--')
    axes[1,0].set_title("Error Scatter Plot")
    axes[1,0].legend()

    # 4. ROC Curve
    axes[1,1].plot(fpr_list, tpr_list, label=f"AUC={roc_auc:.3f}", linewidth=2)
    axes[1,1].plot([0,1], [0,1], linestyle='--')
    axes[1,1].set_title("ROC Curve")
    axes[1,1].legend()

    plt.tight_layout()
    vis_path = f"plots/{att}_vis.png"
    plt.savefig(vis_path)
    plt.close()

    # --- Save text + images into DOCX ---
    report_doc.add_heading(f"Results for {att}", level=1)
    report_doc.add_paragraph(log_buffer.getvalue())
    report_doc.add_picture(cm_path, width=Inches(4.5))
    report_doc.add_picture(vis_path, width=Inches(6))

# Save final doc
report_doc.save("MLP_Final_Attack_Report.docx")
print("Report saved as MLP_Final_Attack_Report.docx")
```

```python
import itertools
import dask.dataframe as dd
import time
import pandas as pd
import numpy as np
import torch
import joblib
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.preprocessing import MinMaxScaler
import copy

#model definition:

import random

class AE(nn.Module):
    def __init__(self, layer_count, neuron_list, input_shape,drop):
        super(AE, self).__init__()

        # Encoder
        layers = []
        for l in range(layer_count - 1):  # each layer connects i  i+1
            if l == 0:
                in_f = input_shape
            else:
                in_f = neuron_list[l]
            out_f = neuron_list[l + 1]
            layers.append(nn.Linear(in_f, out_f))

            if l != len(neuron_list) - 2:
                layers.append(nn.BatchNorm1d(out_f))
                layers.append(nn.ReLU())
                layers.append(nn.Dropout(drop))
        self.encoder = nn.Sequential(*layers)

        # Decoder
        dec_layers = []
        rev = neuron_list[::-1]  # reverse
        for l in range(len(rev) - 1):
            in_f = rev[l]
            if l == len(rev) - 2:   # last layer = output
                out_f = input_shape
                dec_layers.append(nn.Linear(in_f, out_f))
            else:
                out_f = rev[l + 1]
                dec_layers.append(nn.Linear(in_f, out_f))
                dec_layers.append(nn.BatchNorm1d(out_f))
                dec_layers.append(nn.ReLU())
                dec_layers.append(nn.Dropout(drop))
        self.decoder = nn.Sequential(*dec_layers)

    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded
def robust_loss(output, target, delta=1.0):
    residual = torch.abs(output - target)
    return torch.where(
        residual < delta,
        0.5 * residual**2,
        delta * residual - 0.5 * delta**2
    ).mean()
  # Dataset & Model
class NumericDataset(Dataset):
    def __init__(self, data):
        self.data = torch.tensor(data, dtype=torch.float32)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx], self.data[idx]
        #end of model def
#/////////////////////////////////////////////////////////////////////////////
df_big = pd.read_csv("benign-no-label/benign_no_label.csv")


#actual threading:
def threaded_traing(l,n,frac_size,drop,thresh):

  summary=f"number of layers={l},neurons for each layer{n},Drop rate:{drop}fraction size:{frac_size}"


  df = df_big.sample(frac=frac_size, random_state=42).reset_index(drop=True) #Use only 10 percent
  numeric_df = df.select_dtypes(include='number').dropna()

  scaler = MinMaxScaler()   #Scale the data to avoid sensitvity to outliers
  X = scaler.fit_transform(numeric_df)

  # Save the fitted scaler for future use
  joblib.dump(scaler, "base_MLP_scaler.pkl")

  X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)
  #  Training Setup
  train_dataset = NumericDataset(X_train)
  test_dataset = NumericDataset(X_test)
  g = torch.Generator()
  g.manual_seed(42)
  train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True,generator=g)
  test_loader = DataLoader(test_dataset, batch_size=128, shuffle=False)


  device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
  torch.manual_seed(42)
  model=AE(l,n,X.shape[1],drop).to(device)
  total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
  summary=f"Total trainable parameters: {total_params:,}"
  criterion = nn.MSELoss()
  optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
  NUM_EPOCHS = 10
  train_losses = []
  eval_losses = []
  best_eval=99999999
  patience=2
  bad=0
  best_model=None
  #  Training Loop
  for epoch in range(NUM_EPOCHS):
      #time.sleep(0.02)
      model.train()
      total_train_loss = 0
      #loop = tqdm(train_loader, desc=f"Epoch {epoch+1}/{NUM_EPOCHS}", leave=False)
      for batch_x, _ in train_loader:

          batch_x = batch_x.to(device)
          output = model(batch_x)
          loss = robust_loss(output, batch_x)
          optimizer.zero_grad()
          loss.backward()
          optimizer.step()
          total_train_loss += loss.item()
          #loop.set_postfix(loss=loss.item())

      avg_train_loss = total_train_loss / len(train_loader)
      train_losses.append(avg_train_loss)

      # Evaluation
      model.eval()
      total_eval_loss = 0
      with torch.no_grad():
          for batch_x, _ in test_loader:
              batch_x = batch_x.to(device)
              output = model(batch_x)
              loss = criterion(output, batch_x)
              total_eval_loss += loss.item()
      avg_eval_loss = total_eval_loss / len(test_loader)
      eval_losses.append(avg_eval_loss)
      if avg_eval_loss<best_eval :
          #print("yay")
          best_eval=avg_eval_loss
          #best_model = copy.deepcopy(model.state_dict())
      else:
          bad+=1
      if bad>=patience:
          break

    #  Final Evaluation


  model.eval()
  reconstruction_errors = []
  num=0
  with torch.no_grad():
      for batch_x, _ in test_loader:
          num+= 1
          batch_x = batch_x.to(device)
          output = model(batch_x)
          loss = criterion(output, batch_x)  # MSE
          reconstruction_errors.append(loss.cpu().numpy())

  reconstruction_errors = np.array(reconstruction_errors)
  summary+=f"number of batches:{num}"
  #print("number of batches:",num)

  # Threshold from benign (test) distribution
  threshold = np.percentile(reconstruction_errors, thresh)


  mean_error = reconstruction_errors.mean()

  # Classify anomalies (all benign here that are anomalies are actually false positives)
  anomalies = reconstruction_errors > threshold
  false_positives = anomalies.sum()

  # Print results
  summary+=f"Mean reconstruction error: {mean_error:.6f}"
  summary+=f"Anomaly threshold (99th percentile): {threshold:.6f}"
  summary+=f"False positives: {round(false_positives/len(anomalies)* 100)}% out of {len(anomalies)} samples"
  summary+=f"Detected {anomalies.sum()} anomalies out of {len(anomalies)} samples"
  summary+=f"False positives: {round(false_positives/len(anomalies)*100)}% out of {len(anomalies)} samples"
  epochs = range(1, len(train_losses) + 1)
  summary+=f"Epoch {epoch+1}/{NUM_EPOCHS}, Train Loss: {avg_train_loss:.4f}, Eval Loss: {avg_eval_loss:.4f}"
  mean_error = reconstruction_errors.mean()
  std_error = reconstruction_errors.std()
  summary+=f"Mean reconstruction error: {mean_error:.6f}"
  summary+=f"Std reconstruction error:  {std_error:.6f}"
  summary+=f"Anomaly threshold ({threshold}percentile: {threshold:.6f}"
  #/////////////////////////////////////////////////////////////////////////
  #fp testing
  fp_rate=0


  #  Load and Preprocess Malicious & Benign Data

  malicious_df=df_big
  malicious_df = malicious_df.select_dtypes(include='number').dropna()
  malicious_scaled = scaler.transform(malicious_df)

  summary+=f"number of malicious samples:{len(malicious_scaled)}"
  device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
  model.eval()
  malicious_loader = DataLoader(NumericDataset(malicious_scaled), batch_size=128, shuffle=False)
  summary+=f"number of batches:{len(malicious_loader)}"

  #  Get Reconstruction Errors

  def get_reconstruction_errors(loader):
      errors = []
      with torch.no_grad():
          for batch_x, _ in tqdm(loader):
              batch_x = batch_x.to(device)
              output = model(batch_x)

              loss = torch.mean((output - batch_x)**2, dim=1)#MSE
              errors.extend(loss.cpu().numpy())
      return np.array(errors)


  benign_errors = reconstruction_errors
  malicious_errors = get_reconstruction_errors(malicious_loader)

  #  Threshold and Classification
  threshold = np.percentile(benign_errors, thresh)

  benign_preds = benign_errors > threshold
  malicious_preds = malicious_errors > threshold

  # Metrics
  tp = np.sum(malicious_preds)
  fp = np.sum(benign_preds)
  fn = len(malicious_errors) - tp
  tn = len(benign_errors) - fp

  precision = tp / (tp + fp + 1e-8)
  recall = tp / (tp + fn + 1e-8)
  f1 = 2 * precision * recall / (precision + recall + 1e-8)
  accuracy = (tp + tn) / (len(benign_errors) + len(malicious_errors))
  summary+="Phase 2 false positves"
  summary+=f"Precision: {precision:.4f}"
  summary+=f"Recall:    {recall:.4f}"
  summary+=f"F1 Score:  {f1:.4f}"
  summary+=f"Accuracy:  {accuracy:.4f}"
  fp_rate=recall
  #///////////////////////////////////////////////////////////////////////////////
  #testing on all attacks
  attacks=["SSH-Bruteforce.csv","FTP-BruteForce.csv","DoS_attacks-Slowloris.csv","DoS attacks-GoldenEye.csv","DDOS_attack-LOIC-UDP.csv","DDOS_attack-HOIC.csv","Brute_Force_-XSS.csv","Brute_Force_-Web.csv","Bot.csv","sql_injection.csv"]
  for att in attacks:

      df = pd.read_csv('attacks/'+att)

      print(len(df))
      # Drop rows where ' Label' is  'BENIGN'
      df_attack = df[df['Label'] != 'Benign'].copy()

      print(f"Remaining rows: {len(df_attack)}")


      df_attack.to_csv('attack_only2.csv', index=False)
      print("saved")
      df = df.drop(columns=['Label'])
      # Replace inf and -inf with NaN
      import numpy as np
      df = df.replace([np.inf, -np.inf], np.nan)

      # Drop rows with any NaN
      df = df.dropna()

      df.to_csv('attack_no_label2.csv', index=False)
      #########################################

      #  Load and Preprocess Malicious & Benign Data
      malicious_df = pd.read_csv("attack_no_label2.csv")
      malicious_df=malicious_df.drop(columns=["Dst Port","Protocol"])
      malicious_df = malicious_df.select_dtypes(include='number').dropna()
      #malicious_df = malicious_df.sample(frac=0.10, random_state=42).reset_index(drop=True)


      # Load saved scaler from training
      scaler = joblib.load("base_MLP_scaler.pkl")

      # Use the training scaler for both benign & malicious test data
      #benign_scaled = scaler.transform(benign_df)
      malicious_scaled = scaler.transform(malicious_df)

      summary+=f"attack report for:{att}"


      # ------------- Setup -------------
      device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
      #input_dim = malicious_scaled.shape[1]

      # Load model
      #model = ImprovedAutoencoder(input_dim=input_dim).to(device)
      #model.load_state_dict(torch.load("models/regular_best_numeric_autoencoder.pth", map_location=device))

      model.eval()

      # Dataloaders
      #benign_loader = DataLoader(NumericDataset(benign_scaled), batch_size=64, shuffle=False)
      subset_data = malicious_scaled[:min(380, len(malicious_scaled))]
      malicious_loader = DataLoader(NumericDataset(subset_data), batch_size=128, shuffle=False)


      #malicious_loader = DataLoader(NumericDataset(malicious_scaled), batch_size=128, shuffle=False)
      #print(len(malicious_loader))
      #  Get Reconstruction Errors

      def get_reconstruction_errors(loader):
          errors = []
          with torch.no_grad():
              for batch_x, _ in tqdm(loader):
                  batch_x = batch_x.to(device)
                  output = model(batch_x)

                  loss = torch.mean((output - batch_x)**2, dim=1)
                  errors.extend(loss.cpu().numpy())
          return np.array(errors)


      summary+=f"number of malicious samples:{len(subset_data)}"
      summary+=f"number of batches:{len(malicious_loader)}"
      summary+=f"reconstruction error for attack:{att}:{reconstruction_errors}"
      benign_errors = reconstruction_errors
      malicious_errors = get_reconstruction_errors(malicious_loader)

      # Threshold and Classification
      threshold = np.percentile(benign_errors,thresh)

      benign_preds = benign_errors > threshold
      malicious_preds = malicious_errors > threshold

      # Metrics
      tp = np.sum(malicious_preds)
      fp = np.sum(benign_preds)
      fn = len(malicious_errors) - tp
      tn = len(benign_errors) - fp

      precision = tp / (tp + fp + 1e-8)
      recall = tp / (tp + fn + 1e-8)
      f1 = 2 * precision * recall / (precision + recall + 1e-8)
      accuracy = (tp + tn) / (len(benign_errors) + len(malicious_errors))
      f1_scores.append(f1)



  return summary,f1_scores,fp_rate






#end of actual threading
#/////////////////////////////////////////////////////////////////////////////

list_n = []

l_n_p = [i for i in range(1, 13)]       # lengths 1..12
np = [2**i for i in range(12)]          # [2, 4, 8, ..., 4096]
comb_list=[]
for l_n in l_n_p:                       # loop over lengths
  for combo in itertools.combinations_with_replacement(reversed(np), l_n):
    list_n = list(combo)                # keep descending order
    comb_list.append(list_n)
thresholds=[93]
fractions=[0.10]
drop_outs=[0.3]
random.seed(42)          # same subset every run
sample_size = 100
sampled = random.sample(comb_list, sample_size)

results = []
futures=[]
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor, as_completed
import numpy as np
from tqdm import tqdm

# Make your jobs list as before
jobs = []
for sam in sampled:
    for tres in thresholds:
        for frac in fractions:
            for dro in drop_outs:
                jobs.append((len(sam), sam, frac, dro, tres))

list_summary, list_fp_rates, list_f1 = [], [], []
best_f1_so_far = -1
best_summary = None

# Run only 4 at a time
with ThreadPoolExecutor(max_workers=4) as executor:
    futures = [executor.submit(threaded_traing, l, n, frac, dro, tres)
               for (l, n, frac, dro, tres) in jobs]

    for future in tqdm(as_completed(futures), total=len(futures), desc="Training Progress"):
        summary, f1, fp = future.result()
        list_summary.append(summary)
        list_f1.append(f1)
        list_fp_rates.append(fp)

        avg_f1 = np.mean(f1)
        if avg_f1 > best_f1_so_far:
            best_f1_so_far = avg_f1
            best_summary = summary
            with open("best_model_summary.txt", "w") as f:
                f.write("Best Model Summary So Far\n")
                f.write("========================\n\n")
                f.write(best_summary + "\n")
                f.write(f"\nAverage F1: {avg_f1:.4f}\n")
                f.write(f"FP rate: {fp}\n")



```
